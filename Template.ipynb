## Python NumPy For Absolute Beginners






### What is NumPy?
- NumPy stands for Numerical Python
- It is a Python library used for **array** data computations and data manipulations

### Why Learn and Use NumPy?
1. NumPy is faster than traditional Python lists.
2. NumPy is one of the popular Data science libraries. It makes working with numberic data easier and faster.
3. NumPy is interoperable because it supports a wide variety of computer hardware and software.
4. NumPy is an open source library with a very large community for support and assistance.
5. NumPy is easy to learn
